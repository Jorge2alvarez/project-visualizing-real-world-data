```python
# Importar Librerias

import pandas as pd
import numpy as np
import datetime

import seaborn as sns
sns.set()

import plotly as py
#import chart_studio.plotly as py
import cufflinks as cf

cf.go_offline()
```

```python
# Importar Dataset Renfe y crear un Dataframe
data = pd.read_csv('data/renfe.csv')
```

```python
data.head()
```

```python
format_str = '%Y-%m-%d %H:%M:%S' # The format
```

```python
# Formatear el campo fecha como Datatime 
data.start_date = pd.to_datetime(data.start_date)
data.end_date = pd.to_datetime(data.end_date)
```

```python
# Calcular la duracion del viaje
data['trip_duration'] = (pd.to_datetime(data.end_date - data.start_date)).apply(lambda x: x.time())
```

```python
# Categorizar de salida en Morning, Afternoon y Evening
# Crear una nueva columna Time y asignar la hora de salida

data['time'] = pd.to_datetime(data['start_date']).dt.hour

# Convertir la hora de salida en categorias Morning, Afternoon y Evening
data['time'] = np.where(data.time <= 12, 'Morning', 
                        np.where((data.time > 12) & (data.time <= 18), 'Afternoon', 
                        np.where((data.time >= 19), 'Evening', 'X')))
```

```python
data['trip'] = data.origin + ' - ' + data.destination
```

```python
data['weekday'] = data.start_date.apply(lambda x: int(x.weekday()))
```

```python
data['month'] = data.start_date.apply(lambda x: x.month)
```

```python
#data['t_duration_int'] = data.trip_duration.apply(lambda x: x.seconds/60/60)
```

```python
data['start_date_int'] = data.start_date.apply(lambda x: x.time())
```

```python
#data.trip_duration = pd.to_datetime(data.trip_duration)
```

```python
data.head()
```

```python
scater = data.groupby([data.trip, data.start_date_int, data.trip_duration]).price.agg(['mean']).reset_index('trip').reset_index('start_date_int').reset_index('trip_duration')
```

```python
scater = scater.sort_values(by='trip_duration', ascending=False)
```

```python
scater.iplot(x='mean', y='trip_duration', categories='trip',
           xTitle='Price', yTitle='Trip Duration',
           title='Price vs Trip Duration')
```
